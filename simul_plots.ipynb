```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use("seaborn-dark")
```

```python
df = pd.read_csv('measured_data.csv')
usr = df['Users']
resp = np.array(df['Response Time'])
thr = df['Throughput']
cpu = df['Cpu Utilization']
drop = df['Dropped reqs']
```

```python
m_df = pd.read_csv('modeled_data.csv')
resp_m = m_df['Response Time']
thr_m = m_df['Throughput']
cpu_m = 100*m_df['Cpu Utilization']
```

```python
s_r_df = pd.read_csv('response.csv')
user_s = s_r_df['user']
resp_s = 1000*s_r_df['response']
s_u_df = pd.read_csv('utilisation.csv')
util_s = 100*s_u_df['util']
s_t_df = pd.read_csv('throughput.csv')
thr_s = s_t_df['thr']
s_d_df = pd.read_csv('droprate.csv')
drop_s = s_d_df['drop']
```

```python
fig = plt.figure(dpi=360,figsize=(10,5))
plt.xlabel("No. of users",fontweight="bold")
plt.ylabel("Response Time (ms)",fontweight="bold")
plt.title("Simulation Analysis",fontweight="bold")
# plt.xticks(np.arange(0,760,50))
# plt.yticks(np.arange(0,2550,150))
plt.plot(usr, resp, label="Measured", marker='o')
plt.plot(usr, resp_m, label="Modelled", marker='o', color='r')
plt.plot(user_s, resp_s, label="Simulated", color='g')
plt.grid()
plt.legend()
plt.savefig("resp_plot.png")
plt.show()
```

```python
fig = plt.figure(dpi=360, figsize=(10,5))
plt.xlabel("Number of users",fontweight="bold")
plt.ylabel("Throughput (req/sec)",fontweight="bold")
plt.title("Simulation Analysis",fontweight="bold")
# plt.xticks(np.arange(0,750,50))
# plt.yticks(np.arange(0,100,10))
plt.plot(usr, thr, label="Measured", color='b', marker='v')
plt.plot(usr, thr_m, label="Modelled",color='r', marker='v')
plt.plot(user_s, thr_s, label="Simulated",color='g')
plt.grid()
plt.legend()
plt.savefig("thr_plot.png")
plt.show()
```

```python
fig = plt.figure(dpi=360, figsize=(10,5))
plt.xlabel("Number of users",fontweight="bold")
plt.ylabel("Server CPU Utilization",fontweight="bold")
plt.title("Simulation Analysis",fontweight="bold")
plt.xticks(np.arange(0,750,50))
plt.yticks(np.arange(0,110,10))
plt.plot(usr, cpu, label="Measured", color='b', marker='^')
plt.plot(usr, cpu_m, label="Modelled", color='r', marker='^')
plt.plot(user_s, util_s, label="Simulated", color='g')
plt.legend()
plt.grid()
plt.savefig("cpu_plot.png")
plt.show()
```

```python
fig = plt.figure(dpi=360, figsize=(10,5))
plt.xlabel("No of Users")
plt.ylabel("Fraction of requests dropped")
plt.title("Simulation Analysis")
plt.xticks(usr)
plt.plot(usr[:-1], drop[:-1], label="Measured", color='b', marker='s')
plt.plot(user_s, drop_s, label="Simulated", color='g')
plt.grid()
plt.legend()
plt.savefig("drop_plot.png")
plt.show()
```

```python
s_w_df = pd.read_csv('waittime.csv')
wait_s = s_w_df['wait']
user_sw = s_w_df['user']
fig = plt.figure(dpi=360, figsize=(10,5))
plt.xlabel("No of Users")
plt.ylabel("Waiting time (in secs)")
plt.title("Simulation Analysis")
plt.plot(user_sw, wait_s, label="Simulated", color='g')
plt.grid()
plt.legend()
plt.savefig("wait_plot.png")
plt.show()
```
