```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use("seaborn-dark")
```

```python
df = pd.read_csv('resp_conf2.csv')
usr = df['user']
minv = df['min']
maxv = df['max']
fig = plt.figure(dpi=360)
plt.xlabel("No. of users",fontweight="bold")
plt.ylabel("95% confidence intervals",fontweight="bold")
plt.title("Metric: Response time",fontweight="bold")
for i in range(len(usr)):
    plt.plot([usr[i],usr[i]],[minv[i], maxv[i]], color='blue')
plt.grid()
plt.savefig("resp_conf.png")
plt.show()
```

```python
df = pd.read_csv('thr_conf.csv')
usr = df['user']
minv = df['min']
maxv = df['max']
fig = plt.figure(dpi=360)
plt.xlabel("No. of users",fontweight="bold")
plt.ylabel("95% confidence intervals",fontweight="bold")
plt.title("Metric: Throughput",fontweight="bold")
for i in range(len(usr)):
    plt.plot([usr[i],usr[i]],[minv[i], maxv[i]], color='purple')
plt.grid()
plt.xticks()
plt.savefig("thr_conf.png")
plt.show()
```
